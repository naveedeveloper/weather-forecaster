```python
from bs4 import BeautifulSoup
```

```python
from urllib.request import Request, urlopen
```

```python
site = "https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168"

content = Request(site)

page = urlopen(content)

soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup)
```

```python
days = soup.find_all('p', class_ = "period-name")
```

```python
len(days)
```

```python
for day in days:
    print(day.get_text())
```

```python
days_list = list()
for day in days:
    days_list.append(day.get_text())
```

```python
days_list
```

```python
desc = soup.find_all('p', class_ = "short-desc")

desc_list = list()

for desc in desc:
    desc_list.append(desc.get_text())
```

```python
desc_list
```

```python
temp_high = soup.find_all('p', class_ = "temp temp-high")
temp_low = soup.find_all('p', class_ = "temp temp-low")

temp_status_list = list()
temperature_list = list()

for t in temp_high:
    temp_status_list.append(t.string.split(':')[0])
    temperature_list.append(t.string.split(':')[1])
    
for t in temp_low:
    temp_status_list.append(t.string.split(':')[0])
    temperature_list.append(t.string.split(':')[1])
```

```python
temp_status_list
```

```python
temperature_list
```

```python
data = {
    'Day': days_list,
    'Description': desc_list,
    'Temperature Status': temp_status_list,
    'Temperature': temperature_list
}
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```
